```python
from pathlib import Path
import csv

csvpath = Path("Resources/budget_data.csv")

records = []
months = 0
monthly_profit_loss = []
total_profit_loss = 0
difference = 0
last_month_pnl = None
total_difference = 0
min = 0
max = 0
greatest_profit_date = 0
greatest_loss_date = 0

with open(csvpath, "r") as csvfile:
    csvreader = csv.reader(csvfile, delimiter=",")
    csv_header = next(csvreader)
    
    csv_header.append("Changes in PNL")
    records.append(csv_header)
    
    for data in csvreader: 
        months += 1                                 
        monthly_profit_loss = int(data[1])
        total_profit_loss += monthly_profit_loss
       
    #Iterate and find the changes for ((month n+1) - (month n))   
    
        if last_month_pnl == None:
            last_month_pnl = monthly_profit_loss
        else: 
            difference = monthly_profit_loss - last_month_pnl
            last_month_pnl = monthly_profit_loss
        
    #Add Changes in PNL to another Column
        data.append(difference)
        records.append(data)
        
    #Tally the differences in PNL
        difference = data[2]
        total_difference += difference

    #Find the min and max of the differences
        if min == 0:
            min = data[2]
        elif data[2] < min:
            min = data[2]
        elif data[2] > max:
            max = data[2]
   
    #Find month of greatest increase in profits
        if data[2] == max:
            greatest_profit_date = data[0]
     
    #Find month of greatest decrease in losses   
        if data[2] == min:
            greatest_loss_date = data[0] 
    
#Take the average of the total differences in PNL    
    average = round(total_difference/(months-1),2)    
    
#Set the path for the output.txt
output = Path('Resources/budget_data_pnl.txt')

# Open the output path as a file object
with open(output, 'w') as file:
    file.write(f"Financial Analysis\n")
    file.write("---------------------------------------------------\n")
    file.write(f"Total Months: {months}\n")
    file.write(f"Total: ${total_profit_loss}\n")
    file.write(f"Average Change: ${average}\n")
    file.write(f"Greatest Increase in Profits: {greatest_profit_date} (${max})\n")
    file.write(f"Greatest Decrease in Profits: {greatest_loss_date} (${min})\n")
    
```
