```python
#Import the pathlib and csv library
from pathlib import Path
import csv

#Set the file path
csvpath = Path("Resources/budget_data.csv")

#Initialize list of records
records = []

#Define variables
months = 0
monthly_profit_loss = []
total_profit_loss = 0
difference = 0
last_month_pnl = None
total_difference = 0
min = 0
max = 0
greatest_profit_date = 0
greatest_loss_date = 0

#Open the csv file as an object and pass in the csv file to the csv.reader() function 
with open(csvpath, "r") as csvfile:
    
    #Use "," as the delimiter and return the csvreader object 
    csvreader = csv.reader(csvfile, delimiter=",")
    
    #Read the header row
    csv_header = next(csvreader)
    
    #Append the column "Changes in PnL" to the header
    csv_header.append("Changes in PnL")
    
    #Append the new header to the list of records
    records.append(csv_header)
    
    #Read each row of data after the header
    for data in csvreader: 
        
        #Iterate and sum the number of months in the text file
        months += 1  
    
        #Convert the column "Profit/Losses" from string to int
        monthly_profit_loss = int(data[1])
        
        #Iterate and sum the total monthly Profit/Losses
        total_profit_loss += monthly_profit_loss
       
        #Iterate and find the change in PnL for each "Profit/Loss" data line: ((month n+1) - (month n))   
        if last_month_pnl == None:
            last_month_pnl = monthly_profit_loss
        else: 
            difference = monthly_profit_loss - last_month_pnl
            last_month_pnl = monthly_profit_loss
        
        #Append the variable difference to the row of data
        data.append(difference)
    
        #Add the row of data to the list of records. This finalizes the input of data to the "Changes in PnL" column. 
        records.append(data)
         
        #Assign the data in column 2 to the variable differences 
        difference = data[2]
    
        #Iterate and sum the total change in Profit/Losses
        total_difference += difference

        #Find the min and max of the change in Profit/Losses
        if min == 0:
            min = data[2]
        elif data[2] < min:
            min = data[2]
        elif data[2] > max:
            max = data[2]
   
        #Find the date of greatest increase in profits
        if data[2] == max:
            greatest_profit_date = data[0]
     
        #Find the date of greatest decrease in losses   
        if data[2] == min:
            greatest_loss_date = data[0] 
    
    #Take the average of the total changes in PnL    
    average = round(total_difference/(months-1),2)    
    
#Set the path for the output.txt
output = Path('Resources/budget_data_pnl.txt')

#Open the output path as a file object
with open(output, 'w') as file:
    file.write(f"Financial Analysis\n")
    file.write("---------------------------------------------------\n")
    file.write(f"Total Months: {months}\n")
    file.write(f"Total: ${total_profit_loss}\n")
    file.write(f"Average Change: ${average}\n")
    file.write(f"Greatest Increase in Profits: {greatest_profit_date} (${max})\n")
    file.write(f"Greatest Decrease in Profits: {greatest_loss_date} (${min})\n")
```
